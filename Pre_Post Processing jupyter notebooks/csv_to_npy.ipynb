```python
import torch
from torch import Tensor
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pandas as pd
import os
```

```python
def create_df(loc, names):
    
    L = []
    for name in names:
        fp = os.path.join(loc, name + ".csv")
        df = pd.read_csv(fp)
        L.append(df)
    df0 = pd.concat(L)
    return df0

def create_numpys(df):
    L = []
    for i in range(10):
        L.append('year{}_citation_count'.format(i))
    data0 = df[L].to_numpy(dtype=np.float32)
    data  = np.expand_dims(data0, axis=-1)
    labels = df['citationCount'].to_numpy(dtype=np.float32)
    data_size = len(labels)
    
    train_data, rest_data = torch.utils.data.random_split(data, [int(data_size/data_size),data_size - int(data_size/data_size)])
    train_labels, rest_labels = torch.utils.data.random_split(labels, [int(data_size/data_size),data_size - int(data_size/data_size)])
    rest_data_size = len(rest_labels)
    
    valid_data, rest_data2 = torch.utils.data.random_split(rest_data, [int(rest_data_size/rest_data_size),rest_data_size - int(rest_data_size/rest_data_size)])
    valid_labels, rest_labels2 = torch.utils.data.random_split(rest_labels, [int(rest_data_size/rest_data_size),rest_data_size - int(rest_data_size/rest_data_size)])
    rest_data_size2 = len(rest_labels2)
    
    test_data, rest_data2 = torch.utils.data.random_split(rest_data2, [int(rest_data_size2/rest_data_size2),rest_data_size2 - int(rest_data_size2/rest_data_size2)])
    test_labels, rest_labels2 = torch.utils.data.random_split(rest_labels2, [int(rest_data_size2/rest_data_size2),rest_data_size2 - int(rest_data_size2/rest_data_size2)])
    
    np.save('../CitationByYearData/processed3/trainx.npy', train_data)
    np.save('../CitationByYearData/processed3/trainy.npy', train_labels)
    np.save('../CitationByYearData/processed3/validx.npy', valid_data)
    np.save('../CitationByYearData/processed3/validy.npy', valid_labels)
    np.save('../CitationByYearData/processed3/testx.npy', test_data)
    np.save('../CitationByYearData/processed3/testy.npy', test_labels)
        
```
